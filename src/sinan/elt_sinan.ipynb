```python
import pandas as pd
import basedosdados as bd
import os

# ------------------------------------------------------------------
# CONFIGURAÇÕES
# ------------------------------------------------------------------
PROJECT_ID = "barnabe-bq"                          # seu billing_project_id
DEST_DIR   = "/Users/silmara.barnabe/Desktop/dengue/data/staging/sinan"    # pasta‐raiz dos parquet
TABLE      = "basedosdados.br_ms_sinan.microdados_dengue"
CHUNK_MB   = 25                                    # tamanho máx. (MB) de cada arquivo
COMPRESS   = "brotli"                              # brotli, gzip, zstd, snappy…

# ------------------------------------------------------------------
# FUNÇÃO AUXILIAR PARA SALVAR EM VÁRIOS ARQUIVOS PEQUENOS
# ------------------------------------------------------------------
def save_parquet_in_chunks(df: pd.DataFrame, base_path: str,
                           max_size_mb: int = 25, compression: str = "brotli"):
    """
    Salva um DataFrame em arquivos parquet de no máx. `max_size_mb` cada.
    O nome resulta em base_path/part_0000.parquet, part_0001.parquet, …
    """
    # Calcula tamanho aproximado em bytes por linha (amostra das primeiras 1 000 linhas)
    sample_bytes = df.head(1000).memory_usage(index=False, deep=True).sum() / 1000
    rows_per_file = int((max_size_mb * 1024**2) / sample_bytes) or len(df)

    os.makedirs(base_path, exist_ok=True)
    start, part = 0, 0
    while start < len(df):
        end = min(start + rows_per_file, len(df))
        df.iloc[start:end].to_parquet(
            f"{base_path}/part_{part:04d}.parquet",
            compression=compression,
            index=False
        )
        start = end
        part += 1

# ------------------------------------------------------------------
# LOOP SOBRE OS ANOS
# ------------------------------------------------------------------
anos_qt = {
    2000: 172_855, 2001: 488_590, 2002: 897_093, 2003: 416_609, 2004: 136_867,
    2005: 261_501, 2006: 411_022, 2007: 717_097, 2008: 919_324, 2009: 600_658,
    2010: 1_381_254, 2011: 1_150_011, 2012: 950_180, 2013: 2_035_119,
    2014: 966_619, 2015: 2_398_060, 2016: 2_298_020, 2017: 518_483,
    2018: 478_880, 2019: 2_261_956, 2020: 1_495_117, 2021: 1_010_359,
    2022: 1_393_877, 2023: 1_508_653, 2024: 6_442_122, 2025: 1_027_515
}

for ano in anos_qt:
    print(f"⏳  Baixando {ano} ({anos_qt[ano]:,} linhas estimadas)…")
    
    query = f"""
        SELECT *
        FROM `{TABLE}`
        WHERE ano = {ano}
    """

    df = bd.read_sql(query=query, billing_project_id=PROJECT_ID)

    path = f"{DEST_DIR}/microdados_dengue/ano={ano}"
    save_parquet_in_chunks(df, path, CHUNK_MB, compression=COMPRESS)
    print(f"✅  {ano} salvo em {path} ( {len(df):,} linhas )")

print("🎉  Processo concluído!")
```
